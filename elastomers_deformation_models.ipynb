# Deformation and Stress

Deformation refers to the transformation of a body from a certain configuration into a different configuration, usually under an external load. In the case of elastomers, the deformation is recovered after the stress is removed. This type of deformation is referred to as an elastic deformation. 
Homogeneous strain can be defined by three principal extension ratios along three mutually perpendicular axes. Under such strain, a unit cube (Fig. 3.1) is transformed into a parallelepiped having three edge lengths. he extension ratios can be either less than 1, corresponding to compression, or greater than 1, corresponding to stretching, provided that the object being deformed is incompressible (constant volume assumption), that is

$$\lambda_{1} \lambda_{2}\lambda_{2} = 1 $$

is satisfied. Under this assumption, any point on the object represented by the vector $\mathbf{r_{0}}$ with components $(x_{0},y_{0},z_{0})$ in the unstrained state and the same vector $\mathbf{r}$ with components $(x,y,z)$ after deformation can be related by

$$\mathbf{r} = RB\mathbf{r_{0}} $$

<img src="strain&deformationtypes.jpg" alt="strain&deformationtype" width="600"/>

Where $ R = R_{x }R_{y }R_{z }$ is the rotation matrix which describes the deformation angle with respect to the x, y, and z directions, given by the three rotation matrices

$$R_{x} =  \begin{bmatrix} 1&0&0\\ 0&cos\theta&-sin\theta\\0&sin\theta&cos\theta \end{bmatrix} ,  R_{y} =  \begin{bmatrix} cos\theta&0&sin\theta\\ 0&1&0\\-sin\theta&0&cos\theta \end{bmatrix} ,  R_{z} =  \begin{bmatrix} cos\theta&-sin\theta&0\\ sin\theta&cos\theta&0\\0&0&1 \end{bmatrix}$$

and $B$ is the principal stretch matrix, namely 

$$B =  \begin{bmatrix} \lambda_{1}&0&0\\ 0&\lambda_{2}&0\\0&0&\lambda_{3} \end{bmatrix} $$

There are many deformation modes, I will discuss three important modes
1. Uniaxial deformation
2. Equibiaxial deformation
3. Pure shear

In uniaxial deformation, the object is deformed in one direction only (Fig. 3.2a) which corresponds to an increase along the direction of deformation by a ratio $\lambda$ with a corresponding reduction in the other two dimensions. The constant volume assumption then gives

$$\lambda_{1} = \lambda;\qquad     \lambda_{2}=\lambda_{3}=\lambda^{-\frac{1}{2}}$$

In equibiaxial deformation the object is stretched by equal amounts in two perpendicular directions (Fig. 3.2c). If $\lambda_{2}$ is the deformation ratio in the plane of the sheet, the constant volume assumption gives

$$\lambda_{2}=\lambda_{3}=\lambda _{1}^{-\frac{1}{2}}$$

where $\lambda _{1}$ corresponds to the extension ratio in the thickness direction.

In pure shear, the object is elongated in one direction while holding another direction constant. The third direction is perpendicularly shorted according to the constant volume criterion (Fig. 3.2b). If $\lambda_{1}$ is the stretch ratio corresponding to the direction being elongated, and $\lambda_{2}$ is the stretch ratio corresponding to the direction being held constant, the constant volume assumption gives

$$\lambda_{1} = \lambda,\qquad     \lambda_{2}=1 ,\qquad \lambda_{3} =\lambda^{-1}$$

**Deformation function**

We begin by defining the function Deform(lam, t) which takes a stretch vector lam (equivalent to the princple stretch ratio, $\lambda$). The function then outputs three vectors, lam1, lam2 and lam3 which give the deformation in along the 3 princples axes. The variable $t$ defines the deformation type: $t=1$ for uniaxial deformations, $t=2$ for equibiaxial deformations and $t=3$ for pure shear.

```python
#define the function Deform(lam, t)
def Deform(lam, t):
    #uniaxial deformation
    if t == 1:
        lam1 = lam
        lam2 = np.reciprocal(np.sqrt(lam1))
        lam3 = lam2
    #equibiaxial deformation
    elif t == 2:
        lam1 = lam
        lam2 = lam
        lam3 = np.reciprocal(lam**2)
    #pure shear
    elif t == 3:
        lam1 = lam
        lam2 = np.ones((1, lam.size))
        lam3 = np.reciprocal(lam1)
    else: return null
    #return 3xM matrix where M is lam.size
    return np.vstack((lam1,lam2,lam3))   
```



# Macromolecular Structure of Elastomers

The underlying structure of rubber is essentially one of disorganized cluster of randomly oriented, long chain molecules arranged in a network due to cross-linking with very weak intermolecular interactions (Treloar, 2005). Such structure results in a stress-strain relation that is strongly influenced by changes in configurational entropy as the randomly oriented chains become preferentially oriented along a certain direction under deformation. Once the configurational entropy is obtained, statistical mechanics is used to obtain the work and tensile force of the chain from the Helmholtz free energy.

**Elasticity of Gaussian Chains**

In the Gaussian treatment (Treloar, 1944) the probability density, $p(r)$ of the end-to-end length, $r$ for a freely orientated long chain molecule is given by

$$p(r) = 4\pi \left (\frac{3}{2\pi nl^{2}}  \right )^{\frac{3}{2}}  r^{2}exp\left (-\frac{3nr^{2}}{2nl^{2}}  \right )$$

where $n$ is the number of links in the chain and $l$ is the length of each link. The average initial chain length is then given by the root mean square value of $r$

$$r_{rms}=\left (nl^{2}  \right )^{1/2} = \sqrt{n}\;l$$

The configurational entropy $s$ for a single chain is the logarithm of the number of configurations multiplied by the Boltzmann constant, namely

$$s= k\,ln\,p(r)=  c-kb^{2}r^{2}$$

where $c$ is an arbitrary constant, $k$ is the Boltzmann constant and $b^{2}=3/2nl^{2}$.
For a chain whose ends are located at specified points separated by a distance $r$, the work required to move one end of the chain from a distance $r$ to a distance $r+dr$ with respect to the other is equal to the change in the Helmholtz free energy, namely

$$\frac{dW}{dr}=\frac{dA}{dr}=-T\frac{ds}{dr}$$

where $\frac{dW}{dr}=f$ is the tensile force acting along the direction of r. Substitution of the entropy into the force equation gives the tensile force

$$f=2kTb^{2}r$$

Since the tensile force is proportional to $r$, the chain may be regarded an elastic Hookean spring with spring constant $2kTb^{2}$. This result, however, only represents the average value of the tension over a period of time which approaches a constant value as the number of links in the chain is increased. When the number of links is low, continual fluctuations in the location of the chain ends due to its statistical origin will become more obvious. The linear relation is only valid so long as the distance between the two ends of the chain is not so large as to be comparable to its fully extended length. This limitation arises from the Gaussian distribution function from which the force relation is derived. In practice the distance $r$ should not be more than about one third of the fully extended length. For higher extensions, the Gaussian approach becomes increasingly inaccurate and must be replaced by a more suitable distribution function

**Gaussian chain force function**

We define *f_g_chain(T, b, r)*, which takes the inputs T (temperature), b (physical constant) and r (chain length) and outputs the tensile force exterted on the chain.

```python
#define the function f_g_chain(T, b, r)
def f_g_chain(T, b, r):
    k = 1.38e-23            #Boltzmann constant
    return 2*k*T*b*b*r
```

**Elasticity of Gaussian Molecular Networks**

In order to develop a model for rubber elasticity, the chains discussed in the previous section have to be implemented into a network of randomly distributed chains. The solution to this problem (Guth, 1943) gives the following expression for the work of deformation $W$

$$W=\frac{1}{2}NkT(\lambda_{1}^{2}+\lambda_{2}^{2}+\lambda_{3}^{2}-3)$$

Where $N$ is the number of chains per unit volume, $k$ is the Boltzmann constant and $T$ is the absolute temperature.

<img src="gausscomparison.jpg" alt="gausscomparison" width="600"/>

**Gaussian network force function**

We define *f_g_network(G, lam)*, which takes three parameters. The parameter G = NkT is a physical parameter which depends on the material. lam is equivalent to the princple stretch ratio, $\lambda$, which defines the stretching along the main axis (e.g. lam = arange(1, 5, 0.1) defines stretching by 500% along the main axis, with increments of 10%). The variable $t$ defines the deformation type: $t=1$ for uniaxial deformations, $t=2$ for equibiaxial deformations and $t=3$ for pure shear.

```python
#define the function f_g_network(G, lam, t)
def f_g_network(G, lam, t):
    #calculate the three principle stretch ratio vectors
    Lam = Deform(lam, t)
    #seperate the three priciple stretch ratio vectors
    lam1, lam2, lam3 = Lam[0,:], Lam[1,:], Lam[2,:]
    #return the force
    return 0.5*G*(lam1**2 + lam2**2 + lam3**2 - 3)
```

```python
#let's test and plot our function
lam = np.arange(1, 5, 0.1)
G   = 1
f_g = f_g_network(G, lam, 1)

ax = plt.subplot(111)
ax.margins(0,0)
plt.suptitle('  stress-strain relation of Gaussian chain network (uniaxial deformation)')
plt.xlabel('Stretch ratio')
plt.ylabel('Tensile force')
ax.plot(lam, f_g);
```


**Non-Gaussian Chain Statistics**

The Gaussian treatment discussed in the previou section assumes infinitely extensible chains which presents inaccuracies when considering extensions that approach the maximum chain length $(\frac{r}{nl} < ~0.3)$. This limits the usefulness of the Gaussian theory to low stretches (Fig. 3.3).  Langevin chain statistics was used (Kuhn, 1942) to account for the finite extensibility of molecular chains. The distribution function gives the fractional extension ratio as the Langevin function

$$\frac{r}{nl}=\frac{r}{L}=coth(\beta)-\frac{1}{\beta}=L(\beta)$$

Where $n$ is the number of links in the chain and $l$ the length of each segment ($nl$ is the total length of the chain). This gives the probability density, $p(r)$ of the end-to-end length, $r$ in the logarithmic form

$$ln\,p(r)=constant - n\left ( \frac{r}{nl}\beta +ln\frac{\beta }{sinh\,\beta }  \right )$$

The configurational entropy for a single chain follows directly, putting $s=k\,ln\,p(r)$ to obtain

$$s=c - kn\left ( \frac{r}{nl}\beta +ln\frac{\beta }{sinh\,\beta }  \right ) ;\qquad \beta=L^{-1}\left (\frac{r}{nl}  \right )$$

Substitution into the force equation yields on differentiation an expression for the tensile force of a single chain, namely

$$f=\left (\frac{kT}{l}  \right ) L^{-1} \left (\frac{r}{nl}  \right )$$

Which can be expressed in terms of the extension ratio as

$$f(\lambda)=\left (kT\sqrt{n}  \right ) L^{-1} \left (\frac{\lambda}{\sqrt{n}}  \right )$$

**Non-Gaussian chain function**

We first define the inverse langevin function, *InverseLangevin(x)*, which takes an input *x* where *0 < x < 1* and outputs $L^{-1}(x)$ where $L^{-1}(x)$ is the inverse Langevin function. The function uses the [Pade approximant](https://link.springer.com/article/10.1007/BF00366640). 
Next we define *f_ng_chain(lam, T, n)*, the non-Gaussian chain function. *lam* is the principle stretch ratio vector $\lambda$. *T* is the absolute temperature and *n* is the number of links in the chain. The function outputs f, the tensile force exerted on the chain.

```python
#calculate inverse Langevin function
def InverseLangevin(x):
    return x*(3-x**2)/(1-x**2)

#define the non-Gaussian chain force function
def f_ng_chain(lam, T, n):
    k = 1.38e-23            #Boltzmann constant
    #return the force (make sure sqrt(n) > max(lam) to satisfy domain of inverse langevin function)
    return k*T*sqrt(n)*InverseLangevin(lam/sqrt(n)) 
```

```python
#let's test and plot our function
T, n = 300, 100
lam = np.arange(1, 8, 0.1)
f = f_ng_chain(lam, T, n)

ax = plt.subplot(111)
ax.margins(0,0)
plt.suptitle('stress-strain relation of Langevin chain')
plt.xlabel('Stretch ratio')
plt.ylabel('Tensile force')
ax.plot(lam, f);
```

**Non-Gaussian Network Theory**

The problem of incorporating Langevin chains into a full network is considerably more complex than that of Gaussian chains. Unlike the Gaussian network, a simple and exact solution is not possible. To remedy this, many approximative models (Fig. 3.4) have been proposed, with the most widely studied being the 3-chain model, 4-chain model, and 8-chain model.

<img src="differentmodels.jpg" alt="differentmodels" width="600"/>

**3-chain model**

In the 3-chain model (Fig. 3.4a), three chains are located along each of the main axes of the cubic cell. Each chain deforms with the cell, and the stretch on each chain will correspond to a principal stretch value. The resulting work of deformation is given by

$$\begin{split}
W = \frac{NkT}{3}\sqrt{n} \left [  \lambda_{1}\beta_{1} + \sqrt{n}\;ln\left ( \frac{\beta_{1}}{sinh\,\beta_{1}}  \right )\\ + \lambda_{2}\beta_{2} + \sqrt{n}\;ln\left ( \frac{\beta_{2}}{sinh\,\beta_{2}}  \right )\\ + \lambda_{3}\beta_{3} + \sqrt{n}\;ln\left ( \frac{\beta_{3}}{sinh\,\beta_{3}}  \right )  \right ]
\end{split}$$

where $\beta_{i}=L^{-1} \left (\frac{\lambda_{i}}{\sqrt{n}}  \right )$ for $i=1,2,3$. The tensile force can be obtained by differentiating the work function with respect to the principle stretch ratio. 
The 3-chain model shows no state of deformation dependence at high deformations (Fig. 3.5). For the three states of deformation, the stress-strain curves converge whereas the experimental data diverges. The failure to capture this divergence lies in the lack of a cooperative network except that which is imposed by the constant volume criterion, which is only noticable during small deformations.

<img src="3chaindata.jpg" alt="3chaindata" width="600"/>

**4-chain model**

In the 4-chain model (Fig. 3.4b), the four chains start at the center of a tetrahedron and extend towards the four vertices. The tetrahedral structure provides a more cooperative network deformation than the 3-chain model as the chains are allowed to stretch and rotate with the tetrahedron deformation. Unlike the 3-chain model, the chains in the 4-chain model deform even when the deformation does not occur in their respective direction. That is, if the structure was deformed in the x-direction, the y and z chains do not deform in the 3-chain model, while in the 4-chain model, they deform appropriately to reach equilibrium. This means that unlike the 3-chain model, the 4-chain model cannot express the work function analytically. Instead, the relationship between the stretches of the individual chains to the applied stretch is obtained iteratively using numerical methods.
Unlike the 3-chain model, the tetrahedron model shows deformation type dependence at small deformations which is retained at large deformations (Fig. 3.6). This dependence rises due to the cooperative nature of the 4-chain network, where the chain(s) along the deformation axis become locked by the other chains in the network.

<img src="4chaindata.jpg" alt="4chaindata" width="600"/>

**8-chain model**

In the 8-chain model, the chains are located along the diagonals of the unit cell. Due to the models symmetry, the interior joint point remains located in the centre throughout the deformation process. Hence, the stretch on each chain in the structure is found to be the root mean-square of the applied stretches, namely

$$\lambda_{chain}=\sqrt{ \frac{1}{3} (\lambda_{1}^{2}+\lambda_{2}^{2}+\lambda_{3}^{2}) }$$

which results in the simple expression for the strain energy function given by

$$W=NkT\sqrt{n} \left [\beta_{chain}\lambda_{chain}+\sqrt{n}\, ln \left (\frac{\beta_{chain}}{sinh\,\beta_{chain}}  \right )  \right ]$$

where $\beta_{chain}=L^{-1} \left (\frac{\lambda_{chain}}{\sqrt{n}}  \right )$.
In the 8-chain model, the chains undergo stretching for all imposed deformations and also rotate towards the principal axis of stretch, mimicking what would be expected in the deformation of a real network. This results in a highly cooperative network with notable deformation state dependence (Fig. 3.7).

<img src="8chaindata.jpg" alt="8chaindata" width="600"/>

**3-chain and 8-chain force functions**

We define *f_3chain(lam, n, G, t)* and *f_8chain(lam, n, G, t)* as functions to calculate the tensile force exerted on rubber undergoing deformation using the 3-chain and 8-chain model respectively. The variable lam is the principle stretch ratio $\lambda$, *n* is the number of links in the chain, *G* is a physical constant equal to *NkT*, and *t* is the deformation type, with *t=1* for uniaxial deformation, *t=2* for equibiaxial deformation and *t=3* for pure shear.



```python
#define the 3-chain deformation force function
def f_3chain(lam, n, G, t):
    #step size for differentiation
    h = np.mean(np.diff(lam))
    #calculate the 3 principle stretch ratio vectors
    Lam  = Deform(lam, t)
    #calculate the inverse langevin function
    b  = InverseLangevin(Lam/sqrt(n))
    #W1, W2, W3 is the work done on the chains along each axis
    W1 = Lam[0,:]*b[0,:] + sqrt(n)*np.log(b[0,:]/np.sinh(b[0,:]))
    W2 = Lam[1,:]*b[1,:] + sqrt(n)*np.log(b[1,:]/np.sinh(b[1,:]))
    W3 = Lam[2,:]*b[2,:] + sqrt(n)*np.log(b[2,:]/np.sinh(b[2,:]))
    #W_total is the sum of work done on the network
    W_total = (1/3)*G*sqrt(n)*(W1+W2+W3)
    #differentiate W_total to to find force (f = dW/dlam)
    if t == 2: return 0.5*np.diff(W_total)/h 
    else: return np.diff(W_total)/h

#define the 8-chain deformation force function
def f_8chain(lam, n, G, t):
    h = np.mean(np.diff(lam))
    Lam = Deform(lam, t)
    Lam_chain = (1/sqrt(3))*np.sqrt(Lam[0,:]**2 + Lam[1,:]**2 + Lam[2,:]**2)
    #n *= (np.max(Lam_chain)/np.max(lam))**2
    n *= 3/8
    b_chain = InverseLangevin(Lam_chain/sqrt(n))
    W = G*sqrt(n)*((b_chain*Lam_chain) + ((sqrt(n)*np.log(b_chain/np.sinh(b_chain)))))
    if t == 2: return np.diff(W)/(2*h)
    else: return np.diff(W)/h
    
```

```python
#let's test and plot our functions
n, G = 100, 1
lam = np.arange(1,6.5,0.01)
lam0 = lam[:-1]             #force has 1 less element because of differentiation
fig = plt.figure()

#plotting uniaxial deformation stress-strain relation
t = 1
F3 = f_3chain(lam, n, G, t)
F8 = f_8chain(lam, n, G, t)
ax1 = plt.subplot(131)
ax1.margins(0,0)
ax2 = plt.subplot(132)
ax2.margins(0,0)
ax1.plot(lam0, F8, 'k', linewidth=1, label='uniaxial');
ax2.plot(lam0, F3, 'k', linewidth=1, label='uniaxial');
#plotting equibiaxial deformation stress-strain relation
t = 2
F3 = f_3chain(lam, n, G, t)
F8 = f_8chain(lam, n, G, t)
ax1.plot(lam0[:-109], F8[:-109], 'k-.', linewidth=1, label='equibiaxial');
ax2.plot(lam0, F3, 'k-.', linewidth=1, label='equibiaxial');
#plotting pure shear deformation stress-strain relation
t = 3
F3 = f_3chain(lam, n, G, t)
F8 = f_8chain(lam, n, G, t)
ax1.plot(lam0, F8, 'k--', linewidth=1, label='pure shear');
ax2.plot(lam0, F3, 'k--', linewidth=1, label='pure shear');
#formatting title, legend, size...
ax1.legend(loc="upper left")
ax2.legend(loc="upper left")
ax1.title.set_text('stress strain relation (8-chain model)')
ax2.title.set_text('stress strain relation (3-chain model)')
fig.set_figheight(7)
fig.set_figwidth(15)
plt.tight_layout()
```
